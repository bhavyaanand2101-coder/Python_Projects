```python
df = df[['last_updated',
         'temperature_celsius',
         'precip_mm',
         'humidity']]

df = df.fillna(method='ffill').fillna(method='bfill')
df.head()
```

```python
df['last_updated'] = pd.to_datetime(
    df['last_updated'],
    errors='coerce',
    infer_datetime_format=True
)
```

```python
df = df.dropna(subset=['last_updated'])
```

```python
possible_dates = []

for col in df.columns:
    if any(x in col.lower() for x in ["date", "time", "datetime", "update", "epoch", "local"]):
        possible_dates.append(col)

possible_dates
```

```python
df['last_updated'] = pd.to_datetime(df['last_updated'], errors='coerce')
df = df.set_index('last_updated').sort_index()
```

```python
df['date'] = pd.date_range(start='2020-01-01', periods=len(df), freq='H')

df = df.set_index('date')

df.head()
```

```python
df['date'] = pd.date_range(start='2020-01-01', periods=len(df), freq='H')
df = df.set_index('date')
df.head()
```

```python
num_cols = ['temperature_celsius', 'precip_mm', 'humidity']

for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

df[num_cols] = df[num_cols].fillna(method='ffill')
df[num_cols] = df[num_cols].fillna(method='bfill')
```

```python
daily_stats = df[num_cols].resample('D').agg(['mean','min','max','std'])
daily_stats.head()
```

```python
monthly_stats = df[num_cols].resample('M').agg(['mean','min','max','std'])
monthly_stats.head()
```

```python
plt.figure(figsize=(12,4))
plt.plot(df.index, df['temperature_celsius'])
plt.title("Daily Temperature Trend")
plt.xlabel("Date")
plt.ylabel("Temperature (°C)")
plt.show()
```

```python
monthly_rain = df['precip_mm'].resample('M').sum()

plt.figure(figsize=(12,4))
plt.bar(monthly_rain.index, monthly_rain.values)
plt.title("Monthly Rainfall Total (mm)")
plt.xlabel("Month")
plt.ylabel("Rainfall")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6,5))
plt.scatter(df['temperature_celsius'], df['humidity'], alpha=0.5)
plt.title("Humidity vs Temperature")
plt.xlabel("Temperature (°C)")
plt.ylabel("Humidity (%)")
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(12,4))

ax1.plot(df.index, df['temperature_celsius'], color='red')
ax1.set_ylabel("Temperature (°C)")

ax2 = ax1.twinx()
ax2.bar(monthly_rain.index, monthly_rain.values, alpha=0.3)
ax2.set_ylabel("Rainfall (mm)")

plt.title("Temperature Trend + Monthly Rainfall")
plt.show()
```

```python
df['month'] = df.index.month
df.groupby('month')[num_cols].mean()
```

```python
def season(m):
    if m in [12,1,2]: return "Winter"
    if m in [3,4,5]: return "Spring"
    if m in [6,7,8]: return "Summer"
    return "Autumn"

df['season'] = df.index.month.map(season)
df.groupby('season')[num_cols].mean()
```
